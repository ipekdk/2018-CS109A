```
from elsapy.elsclient import ElsClient
from elsapy.elsprofile import ElsAuthor, ElsAffil
from elsapy.elsdoc import FullDoc, AbsDoc
from elsapy.elssearch import ElsSearch
import json
```

```
## Initialize client
client = ElsClient("7f69d555b3ec651f503ccf5ba1415217")
```

```
doc_srch = ElsSearch("ABS(artificial intelligence) AND ABS(health) AND DOCTYPE(ar) AND PUBYEAR = 2020",'scopus')
doc_srch.execute(client, get_all = True)
print ("doc_srch has", len(doc_srch.results), "results.")
```

```
myNewArray = []
for i in range(len(doc_srch.results)):
  if ("prism:doi" in doc_srch.results[i]):
    myNewArray.append(doc_srch.results[i]['prism:doi'])
    print(i)
    print(doc_srch.results[i]['prism:doi'])
```

```
import time
import urllib.request, json 

myFinalArray = []

for i in range(len(myNewArray)):
  print(myNewArray[i])
  doi_doc = FullDoc(doi = myNewArray[i])
  if doi_doc.read(client):
      print ("doi_doc.title: ", doi_doc.title)
      doi_doc.write()   
      print ("doi_doc.description: ", doi_doc.data["coredata"]["dc:description"])
      myFinalArray.append({"title": doi_doc.title,"description": doi_doc.data["coredata"]["dc:description"],"doi":doi_doc.data["coredata"]["prism:doi"]})
  else:
      print ("Read document failed.")
  time.sleep(0.1)
```
